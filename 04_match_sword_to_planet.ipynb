```python
import geopandas as gpd
from glob import glob
from tqdm import tqdm
import pandas as pd
import numpy as np
import os
```

```python
sword_dir = "/nas/cee-water/cjgleason/data/SWORD/SWORDv16/shp"
subdirs = ["AF", "AS", "EU", "NA", "OC", "SA"]
```

```python
all_sword_fps = []
for subdir in subdirs:
    for fp in glob(os.path.join(sword_dir, subdir, "*reaches*.shp")):
        all_sword_fps.append(fp)
        print(fp)
print(len(all_sword_fps))
```

```python
# Create mapping of reach_id to sword filepath
all_sword_mapping = None
for fp in tqdm(all_sword_fps):
    data = gpd.read_file(fp)
    mapping_data = data[['reach_id', 'reach_len','width','width_var']]
    mapping_data["sword_fp"] = fp
    if all_sword_mapping is None:
        all_sword_mapping = mapping_data
    else:
        all_sword_mapping = pd.concat([all_sword_mapping, mapping_data])
all_sword_mapping.to_csv("all_sword_mapping_v16.csv", index=False)  
```
